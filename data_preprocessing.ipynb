Import bibliotek

```python
import numpy as np
```

```python
import pandas as pd
import sklearn

sklearn.__version__
```

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'male'],
    'price': [199.0, 89.0, 99, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}


df_raw = pd.DataFrame(data=data)
df_raw
```

```python
df = df_raw.copy()
df.info()
```

```python
for col in ['size', 'color', 'gender', 'bought']:
  df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')

df.info()
```

```python
df.describe()
```

```python
df.describe().T
```

```python
df.describe(include=['category']).T
```

2. LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder # klasa używana do przekształcania zmiennej wynikowej

le = LabelEncoder() # tworzymy instancję klasy
le.fit(df['bought']) # dopasowuje encoder do naszej zmiennej
le.transform(df['bought']) # metoda transformuje dane; zwrócone 0 i 1;
```

```python
le.fit_transform(df['bought']) # metod fit i transform nie trzeba stosować oddzielnie
```

```python
le.classes_ # wyświetla mapowanie (która kategoria jakiej liczbie odpowiada)
```

```python
df['bought'] = le.fit_transform(df['bought']) # zamieniamy to, co jest w df
```

```python
df
```

```python
le.inverse_transform(df['bought']) # można wrócić do stanu sprzed transformacji
```

```python
df['bought']=le.inverse_transform(df['bought'])
```

```python
df
```

```python
df['bought'] = le.fit_transform(df['bought'])
```

```python
df
```

3. OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder # importujemy klasę; utworzy k zmiennych gdzie k to liczba kategorii zmiennej
```

```python
encoder = OneHotEncoder(sparse = False) # tworzymy instancję, ale nie macierz rzadką
```

```python
encoder.fit(df[['size']])
```

```python
encoder.transform(df[['size']])
```

```python
encoder.categories_ # mówi, co jest w kolumnach
```

```python
encoder = OneHotEncoder(drop='first', sparse=False) # nowa instancja, robi dummies dla wybranej zmiennej z pierwszą kategorią jako kategorią odniesienia
```

```python
encoder.fit(df[['size']])
```

```python
encoder.transform(df[['size']])
```

4. Pandas *get_dummies()*

5. Standaryzacja Danych

std() - pandas nieobciążony

std() - numy obciążony

```python
print(f"{df['price']}/n")
print(f"Średnia: {df['price'].mean()}")
print(f"Odchylenie standardowe: {df['price'].std():.4f}")
```

```python
(df['price']-df['price'].mean())/df['price'].std()
```

```python
def standardize(x):
  return (x-x.mean())/x.std()

standardize(df['price'])
```

```python
from sklearn.preprocessing import scale
```

```python
scale(df['price'])
```

```python
from sklearn.preprocessing import StandardScaler
```
